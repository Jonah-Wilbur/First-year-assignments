```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = smp.symbols('x y')
```

```python
f = smp.exp(x)*smp.sin(y)+y**3
f
```

```python
g = x**2*y + x*y**2
g
gradient = smp.Matrix([smp.diff(f, x), smp.diff(f, y)]) #define the gradient using symbolic caluclations, will print as a matrix so it is clear
gradient
```

```python
def grad(x, y):
    return ((np.exp(x)*np.sin(y))**2 + (3*y**2 + np.exp(x)*np.cos(y))**2)**0.5 #This is the function shown above, the first partials of the given function
print(grad(1,-1))
```

```python
h = smp.ln(x**2+y**2)
h
dh_dx = smp.diff(h, x)
dh_dy = smp.diff(h, y)
d2h_dx2 = smp.diff(dh_dx, x)
d2h_dy2 = smp.diff(dh_dy, y)
d2h_dxdy = smp.diff(dh_dx, y)
d2h_dydx = smp.diff(dh_dy, x)
hess = smp.Matrix([[d2h_dx2,d2h_dxdy],[d2h_dydx,d2h_dy2]]) #This is just using more symbolic caluclus to determine the derivatives for all of these expressions, defined in a matrix
hess #print the hessian matrix
```

The mixed partial derivatives are symmetrical, the fucntion is nice enough that the mixed second partials have to be identical

```python
j = x**3 -3*x*y + y**3
j_func = smp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400) #set the x values for the plot
y_vals = np.linspace(-3, 3, 400) #set the y values for the plot
X, Y = np.meshgrid(x_vals, y_vals) #mesh them together, to print z well
Z = j_func(X, Y) #define z as f(x, y)
plt.contourf(X, Y, Z, levels=50, cmap='viridis') #plot the values against each other, with a set number (50) of different colours
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
q = smp.sin(x**2+y**2) #define q (equivalent to z) as a sin (x^2 +y^2)
q_func = smp.lambdify((x, y), q)
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = q_func(X, Y) #my own function I made
plt.contourf(X, Y, Z, levels=50, cmap='viridis') #print the contour plot, with a set number of co
plt.colorbar()
plt.title('Contour plot of $j(x, y) = sin(x^2+y^2)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
